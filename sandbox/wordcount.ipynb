```python

source_path = "C:\\Users\\victo\\PycharmProjects\\WordCountSpark\\pyspark_impacta\\datasets\\lorem_ipsum\\lorem_ipsum.txt"

rdd = spark.sparkContext.textFile(source_path)
rdd.collect()
```

```python
rdd1 = rdd.filter(lambda x: x.replace(',', ''))

rdd2 = rdd1.flatMap(lambda x: x.split(" ")).map(lambda x: (x, 1))

rdd3 = rdd2.reduceByKey(lambda x, value : x + value).sortBy(lambda x: x[-1], ascending = False)

rdd3.collect()

```
